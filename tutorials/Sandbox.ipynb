# Sandbox for C Types


```python
%matplotlib widget
# Uncomment for interactive plots when running the notebook!
import numpy as np

import PyPO
import PyPO.Structs

import ctypes
```

```python
inp = np.arange(0, 100.0).reshape((10,10))
```

```python
inp
```

```python
out = PyPO.Structs.reflcontainer()
```

```python
out.x = (ctypes.c_double * inp.size)(*inp.ravel().tolist())
```

```python
out.x
```

```python
outp = np.ctypeslib.as_array(out.x, shape=inp.shape).astype(np.float32)
```
